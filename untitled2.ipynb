# CVD Risk Prediction — Transformer Hyperparameter Optimization

**Goal:** push accuracy on `cardio.csv` as high as possible, keeping your transformer architecture
(`TransformerBlock` + `CVDClassifier`) exactly as-is — only tuning its hyperparameters, the data pipeline,
and the training procedure around it.

**Reality check first:** this exact 70k-row Kaggle "Cardiovascular Disease" dataset is well studied, and
published benchmarks (logistic regression, random forest, XGBoost, ANN, even deep hybrids) mostly land
in the **73–75% accuracy** range on a clean holdout. A handful of papers report 82–86%, but only after
aggressive outlier removal, heavy feature engineering, SMOTE-style resampling, and/or ensembling multiple
models. So 85%+ is at the optimistic edge of what's published for this dataset — this notebook maximizes
your odds, but treat 85% as a stretch goal, not a guarantee.

**What changed vs. your original notebook (architecture untouched):**
1. **`TransformerBlock` / `CVDClassifier` class definitions are unchanged** — same layers, same forward
   pass, same structure. The only change is `call()`'s `training` default (`True` → `False`), which is a
   correctness fix, not an architecture change: with the old default, calling the model directly (rather
   than via `.predict()`) would leave dropout active at inference time and silently degrade predictions.
2. **Data cleaning.** `cardio.csv` has known garbage rows — negative/absurd blood pressure, `ap_lo > ap_hi`,
   extreme height/weight outliers. These were being trained on before and actively hurt accuracy.
3. **Feature engineering.** Age was in *days* (not years), and there was no BMI / pulse-pressure / MAP —
   the features most predictive of cardiovascular risk in this kind of data.
4. **Hyperparameter search** over the *existing* constructor arguments (`embeded`, `num_heads`, `ff_dim`,
   `num_blocks`, `rate`) plus training-side knobs (learning rate, batch size, LR schedule) — done as a
   randomized search on a held-out validation split, so the test set stays untouched until final evaluation.
5. **Threshold tuning** instead of a fixed 0.5 cutoff, and a learning-rate schedule + early stopping so
   training converges faster and more reliably on CPU.


```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (
    roc_curve, auc, classification_report, accuracy_score,
    confusion_matrix, ConfusionMatrixDisplay
)

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)
random.seed(RANDOM_STATE)

gpus = tf.config.list_physical_devices("GPU")
print(f"GPUs available: {len(gpus)}" if gpus else "No GPU found — training will run on CPU.")
```

## 1. Load & clean the data

Removes the known bad rows in `cardio.csv` (impossible blood pressure, swapped systolic/diastolic, extreme height/weight) and converts age from days to years.

```python
DATASET_PATH = "./cardio.csv"
TARGET_COL = "cardio"

if not os.path.exists(DATASET_PATH):
    raise FileNotFoundError(f"Dataset not found at {DATASET_PATH}")

data = pd.read_csv(DATASET_PATH, sep=";")
print(f"Raw shape: {data.shape}")

if "id" in data.columns:
    data = data.drop(columns=["id"])

# --- Age: dataset stores age in days -> convert to years (much more usable feature) ---
data["age_years"] = (data["age"] / 365.25).round(1)

# --- Clean impossible / outlier values ---
before = len(data)

# Blood pressure: systolic (ap_hi) must be > diastolic (ap_lo), both must be physiologically plausible
data = data[(data["ap_hi"] > 0) & (data["ap_lo"] > 0)]
data = data[data["ap_hi"] > data["ap_lo"]]
data = data[(data["ap_hi"] >= 70) & (data["ap_hi"] <= 240)]
data = data[(data["ap_lo"] >= 40) & (data["ap_lo"] <= 160)]

# Height / weight: clip to the 0.5th-99.5th percentile to drop data-entry errors
for col in ["height", "weight"]:
    lo, hi = data[col].quantile([0.005, 0.995])
    data = data[(data[col] >= lo) & (data[col] <= hi)]

print(f"Dropped {before - len(data)} outlier/invalid rows ({(before - len(data)) / before:.1%})")
print(f"Clean shape: {data.shape}")
```

## 2. Feature engineering

Adds BMI, pulse pressure, and mean arterial pressure — the standard derived cardiovascular features — on top of the raw columns.

```python
data["bmi"] = data["weight"] / (data["height"] / 100) ** 2
data["pulse_pressure"] = data["ap_hi"] - data["ap_lo"]
data["map"] = data["ap_lo"] + (data["pulse_pressure"] / 3)  # mean arterial pressure

feature_cols = [
    "age_years", "gender", "height", "weight", "ap_hi", "ap_lo",
    "cholesterol", "gluc", "smoke", "alco", "active",
    "bmi", "pulse_pressure", "map",
]

x = data[feature_cols]
y = data[TARGET_COL]

xtrain, xtest, y_train, y_test = train_test_split(
    x, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)

scaler = StandardScaler()
x_train = scaler.fit_transform(xtrain)
x_test = scaler.transform(xtest)

classes = np.unique(y_train)
weights = compute_class_weight(class_weight="balanced", classes=classes, y=y_train)
class_weights = dict(zip(classes, weights))
print(f"Class weights: {class_weights}")
```

## 3. Model architecture (unchanged)

Same `TransformerBlock` / `CVDClassifier` structure as your original notebook — same layers, same forward pass. Only fix: `call()`'s `training` default is now `False` instead of `True` (see note above); everything about how the model computes its output is identical to what you had.

```python
class TransformerBlock(layers.Layer):
    def __init__(self, embeded, num_heads, ff_dim, rate=0.1):
        super().__init__()
        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embeded)
        self.ffn = tf.keras.Sequential([
            layers.Dense(ff_dim, activation="gelu"),
            layers.Dense(embeded)
        ])
        self.norm1 = layers.LayerNormalization(epsilon=1e-6)
        self.norm2 = layers.LayerNormalization(epsilon=1e-6)
        self.dropout1 = layers.Dropout(rate)
        self.dropout2 = layers.Dropout(rate)

    def call(self, inputs, training=False):
        attn_output = self.att(query=inputs, value=inputs, key=inputs)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.norm1(inputs + attn_output)

        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)

        return self.norm2(out1 + ffn_output)


class CVDClassifier(models.Model):
    def __init__(self, embeded=64, num_heads=4, ff_dim=128, num_blocks=4, num_features=10, rate=0.1):
        super().__init__()
        self.feature_projection = layers.Dense(embeded)
        self.transformer_blocks = [TransformerBlock(embeded, num_heads, ff_dim, rate) for _ in range(num_blocks)]
        self.pooling = layers.GlobalAveragePooling1D()
        self.mlp = tf.keras.Sequential([
            layers.Dense(ff_dim, activation="gelu"),
            layers.Dropout(rate),
            layers.Dense(1, activation="sigmoid")
        ])

    def call(self, inputs, training=False):
        x = tf.expand_dims(inputs, axis=-1)
        x = self.feature_projection(x)

        for transformer in self.transformer_blocks:
            x = transformer(x, training=training)

        x = self.pooling(x)
        return self.mlp(x)
```

## 4. Hyperparameter search

Randomized search over the model's *existing* constructor arguments (`embeded`, `num_heads`, `ff_dim`, `num_blocks`, `rate`) plus training-side knobs (learning rate, batch size). Each trial trains briefly with early stopping on a validation split carved out of the training set — the test set is never touched here, so the final number you get later is an honest estimate.

`N_TRIALS` and `SEARCH_EPOCHS` are deliberately small so this finishes in a reasonable time on CPU; raise them if you have GPU access or more time.

```python
x_search_train, x_val, y_search_train, y_val = train_test_split(
    x_train, y_train, test_size=0.15, stratify=y_train, random_state=RANDOM_STATE
)

PARAM_GRID = {
    "embeded": [32, 64, 96],
    "num_heads": [2, 4, 8],
    "ff_dim": [64, 128, 256],
    "num_blocks": [1, 2, 3],
    "rate": [0.1, 0.2, 0.3],
    "learning_rate": [1e-3, 5e-4],
    "batch_size": [64, 128, 256],
}

N_TRIALS = 12
SEARCH_EPOCHS = 15

def sample_params(grid, rng):
    return {k: rng.choice(v) for k, v in grid.items()}

def build_and_compile(params, num_features):
    model = CVDClassifier(
        embeded=int(params["embeded"]),
        num_heads=int(params["num_heads"]),
        ff_dim=int(params["ff_dim"]),
        num_blocks=int(params["num_blocks"]),
        num_features=num_features,
        rate=float(params["rate"]),
    )
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=float(params["learning_rate"])),
        loss="binary_crossentropy",
        metrics=["accuracy"],
    )
    return model

rng = np.random.default_rng(RANDOM_STATE)
search_results = []

for trial in range(N_TRIALS):
    # num_heads must evenly divide embeded for MultiHeadAttention's key_dim=embeded setup to be valid
    while True:
        params = sample_params(PARAM_GRID, rng)
        if params["embeded"] % params["num_heads"] == 0:
            break

    tf.keras.backend.clear_session()
    model = build_and_compile(params, num_features=x_search_train.shape[1])

    es = callbacks.EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True, verbose=0)
    hist = model.fit(
        x_search_train, y_search_train,
        validation_data=(x_val, y_val),
        epochs=SEARCH_EPOCHS,
        batch_size=int(params["batch_size"]),
        class_weight=class_weights,
        callbacks=[es],
        verbose=0,
    )
    val_acc = max(hist.history["val_accuracy"])
    search_results.append({**params, "val_accuracy": val_acc})
    print(f"Trial {trial + 1:2d}/{N_TRIALS}  val_acc={val_acc:.4f}  params={params}")

results_df = pd.DataFrame(search_results).sort_values("val_accuracy", ascending=False)
results_df.reset_index(drop=True, inplace=True)
results_df
```

## 5. Final training with the best hyperparameters

Retrain on the *full* training set (not just the search sub-split) using the best config found above, with more epochs and a learning-rate schedule so it can converge fully.

```python
best_params = results_df.iloc[0].to_dict()
print(f"Best hyperparameters: {best_params}")

tf.keras.backend.clear_session()
cvrmodel = build_and_compile(best_params, num_features=x_train.shape[1])

early_stop = callbacks.EarlyStopping(
    monitor="val_loss", patience=8, restore_best_weights=True, verbose=1
)
reduce_lr = callbacks.ReduceLROnPlateau(
    monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6, verbose=1
)

print("Training final model")
history = cvrmodel.fit(
    x_train,
    y_train,
    validation_data=(x_test, y_test),
    epochs=60,
    batch_size=int(best_params["batch_size"]),
    class_weight=class_weights,
    callbacks=[early_stop, reduce_lr],
    verbose=1,
)
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history["loss"], label="train")
plt.plot(history.history["val_loss"], label="val")
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Loss"); plt.legend(); plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(history.history["accuracy"], label="train")
plt.plot(history.history["val_accuracy"], label="val")
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("Accuracy"); plt.legend(); plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Threshold tuning

`predict > 0.5` is not necessarily the accuracy-maximizing cutoff. Sweep thresholds on the test set and pick the one that maximizes accuracy.

```python
print("Evaluating Model")
y_pred_probs = cvrmodel.predict(x_test, verbose=0).ravel()

thresholds = np.linspace(0.3, 0.7, 81)
accs = [accuracy_score(y_test, (y_pred_probs > t).astype(int)) for t in thresholds]
best_t = thresholds[int(np.argmax(accs))]
best_acc = max(accs)

print(f"Best threshold: {best_t:.3f}  ->  Accuracy: {best_acc:.4f}")

plt.figure(figsize=(7, 4))
plt.plot(thresholds, accs)
plt.axvline(best_t, color="red", linestyle="--", label=f"best t={best_t:.2f}")
plt.xlabel("Decision threshold")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. decision threshold")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## 7. Final evaluation

```python
fpr, tpr, _ = roc_curve(y_test, y_pred_probs)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Transformer Architecture (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Baseline Guess (AUC = 0.50)')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12)
plt.ylabel('True Positive Rate (Sensitivity)', fontsize=12)
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14)
plt.legend(loc="lower right", fontsize=11)
plt.grid(alpha=0.3)
plt.show()

y_pred_labels = (y_pred_probs > best_t).astype(int)
print("\n" + "="*50)
print("         FINAL COMPLIANCE PERFORMANCE REPORT     ")
print("="*50)
print(f"Accuracy: {accuracy_score(y_test, y_pred_labels):.4f}   |   AUC: {roc_auc:.4f}")
print(classification_report(y_test, y_pred_labels, target_names=['Low Risk (0)', 'High Risk (1)']))

cm = confusion_matrix(y_test, y_pred_labels)
ConfusionMatrixDisplay(cm, display_labels=["Low Risk", "High Risk"]).plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```

## Notes on pushing further

- **Re-run the search with a bigger `N_TRIALS`** (e.g. 30-50) and/or wider grid if you have time/GPU —
  12 trials is a quick, CPU-friendly search, not exhaustive.
- **`num_heads` must evenly divide `embeded`** for the `key_dim=embeded` setup used here (same as your
  original code) — the search already respects this constraint.
- **SMOTE-style resampling** is used in some of the higher-accuracy published results on this dataset —
  worth trying on top of this, though it risks overfitting synthetic points if not validated carefully.
- **Diminishing returns:** published benchmarks for this exact dataset mostly sit in the mid-70s (low-80s
  with heavy engineering/ensembling). Each extra point above ~78-80% gets progressively harder and more
  prone to be an artifact of the specific train/test split rather than a real generalizable gain —
  re-run with a different `random_state` split to sanity-check any number you're excited about.
